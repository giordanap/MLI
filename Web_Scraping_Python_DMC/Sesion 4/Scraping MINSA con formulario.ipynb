```python
driver = Chrome('chromedriver_win32/chromedriver.exe')
link = 'http://observatorio.digemid.minsa.gob.pe/'
driver.get(link)

content = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
medicamento = 'paracetamol'

driver.implicitly_wait(10)
driver.find_element_by_id('txtBuscador').send_keys(medicamento)
```

```python
autocompletado = driver.find_elements_by_class_name('ui-menu-item')
autocompletado[0].click()
n_meds = len(autocompletado)
driver.find_element_by_id('btnBuscar').click()
```

```python
content = BeautifulSoup(driver.find_element_by_class_name('dataTables_scroll').get_attribute('innerHTML'), 'html.parser')
titulos = content.find(class_='dataTables_scrollHead').find('tr').select('th')

dicts = []
keys = []
for tit in titulos:
    keys.append(tit.text)

rows = content.find(class_='dataTables_scrollBody').find(id='tbPrivados').select('tr')

for row in rows:
    data = []
    cols = row.select('td')
    for col in cols:
        data.append(col.text)
    dicts.append(dict(zip(keys,data)))
```

```python
for i in range(len(autocompletado)-1):
    driver.find_element_by_class_name('ui-autocomplete-input').clear()
    driver.implicitly_wait(10)
    driver.find_element_by_class_name('ui-autocomplete-input').send_keys(medicamento)
    driver.find_elements_by_class_name('ui-menu-item')[i+1].click()
    driver.find_element_by_name('btn_buscar').click()
    
    rows = content.find(class_='dataTables_scrollBody').find(id='tbPrivados').select('tr')

    for row in rows:
        data = []
        cols = row.select('td')
        for col in cols:
            data.append(col.text)
        dicts.append(dict(zip(keys,data)))
```
